```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

#Charger les données à partir du fichier CSV
data = pd.read_csv("donnees_vin.csv")
```

```python
#Tracer les histogrammes des variables

fig, axes = plt.subplots(figsize=(19, 10),nrows=3, ncols=4)  # Create a new figure with a 3x4 grid of subplots

# Assuming 'data' is a DataFrame or array-like object containing your data
for ax, column in zip(axes.flatten(), data.columns):
    data[column].hist(ax=ax, alpha=0.5, label='x', color='blue')  # Plot the histogram on each subplot

plt.show()  # Display the figure
```

```python
#Calculer la matrice de corrélation entre les variables
correlation_matrix = data.corr()

#Afficher la matrice de corrélation
#print(correlation_matrix)

data.corr()
```

```python
#Sélectionner les caractéristiques à utiliser
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

#Séparer les caractéristiques (X) et la variable cible (Y)
X = data[features]
Y = data['quality']
```

```python
#Diviser les données en ensembles d'entraînement et de test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

#Créer une instance du modèle de régression linéaire
model = LinearRegression()

#Entraîner le modèle sur l'ensemble d'entraînement
model.fit(X_train, Y_train)

#Prédire la qualité du vin sur l'ensemble de test
Y_pred = model.predict(X_test)
```

```python
#Calculer l'erreur quadratique moyenne (MSE)
mse = mean_squared_error(Y_test, Y_pred)

#Calculer le coefficient de détermination (R2)
r2 = r2_score(Y_test, Y_pred)
#Afficher les métriques de performance
print("Erreur quadratique moyenne (MSE) :", mse)
print("Coefficient de détermination (R2) :", r2)

```

```python
#Afficher le graphe de dispersion des valeurs prédites et réelles
plt.scatter(Y_test, Y_pred)
plt.xlabel("Valeurs réelles")
plt.ylabel("Valeurs prédites")
plt.title("Graphe de dispersion des valeurs réelles et prédites")
plt.show()
```
